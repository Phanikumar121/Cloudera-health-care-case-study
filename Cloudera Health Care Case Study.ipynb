# Part 1

```python
#1A)Highest cost variation for in patient
Medicare_VAR = Medicare_Provider_Inpatient[['DRG Definition','Average Total Payments']].groupby(['DRG Definition']).agg('std').add_suffix('SD')
Medicare_VAR1 = Medicare_Provider_Inpatient[['DRG Definition','Average Total Payments']].groupby(['DRG Definition']).agg('mean').add_suffix('Mean')
Medicare_VAR['Mean_Average Total Payments'] = Medicare_VAR1['Average Total PaymentsMean']
```

```python
#Variance is calculated by dividing standard deviation with mean
Medicare_VAR['Variation'] = Medicare_VAR.apply(lambda x: x['Average Total PaymentsSD']/x['Mean_Average Total Payments'],axis =1)
```

```python
pd.DataFrame(Medicare_VAR['Variation'].nlargest(11)).reset_index()
```

```python
#1A)Highest cost variation for out patient
Medicare_VAR = Medicare_Provider_Outpatient[['APC','Average Total Payments']].groupby(['APC']).agg('std').add_suffix('SD')
Medicare_VAR1 = Medicare_Provider_Outpatient[['APC','Average Total Payments']].groupby(['APC']).agg('mean').add_suffix('Mean')
Medicare_VAR['Mean_Average Total Payments'] = Medicare_VAR1['Average Total PaymentsMean']
```

```python
#Variance is calculated by dividing standard deviation with mean
Medicare_VAR['Variation'] = Medicare_VAR.apply(lambda x: x['Average Total PaymentsSD']/x['Mean_Average Total Payments'],axis =1)
```

```python
pd.DataFrame(Medicare_VAR['Variation'].nlargest(11)).reset_index()
```

```python
#1B)Highest-Cost Claims by Provider for in patient
q1b = Medicare_Provider_Inpatient.groupby(by = ['Provider Name'])[['Average Medicare Payments']].sum().add_prefix("Sum_of_").round(1)
q1b.sort_values(by = ['Sum_of_Average Medicare Payments'],ascending = False).head().reset_index()
```

```python
#1B)Highest-Cost Claims by Provider for out patient
q1b2 = Medicare_Provider_Outpatient.groupby(by = ['Provider Name'])[['Average Total Payments']].sum().add_prefix("Sum_of_").round(1)
q1b2.sort_values(by = ['Sum_of_Average Total Payments'],ascending = False).head().reset_index()
```

```python
#1C)Highest-Cost Claims by Region for in patient
q1c = Medicare_Provider_Inpatient.groupby(by = ['Provider State'])[['Average Medicare Payments']].sum().add_prefix("Sum_of_").round(1)
q1c.sort_values(by = ['Sum_of_Average Medicare Payments'],ascending = False).head().reset_index()
```

```python
#1C)Highest-Cost Claims by Region for out patient
q1c2 = Medicare_Provider_Outpatient.groupby(by = ['Provider State'])[['Average Total Payments']].sum().add_prefix("Sum_of_").round(1)
q1c2.sort_values(by = ['Sum_of_Average Total Payments'],ascending = False).head().reset_index()
```

```python
#Part1D) Highest Number of Procedures and Largest Differences between Claims and Reimbursements for in patient
Medicare_Provider_Inpatient['Diff b/w Claims and Reimb'] = Medicare_Provider_Inpatient['Average Covered Charges'] - Medicare_Provider_Inpatient['Average Total Payments']
q1d = Medicare_Provider_Inpatient.groupby(by = ['DRG Definition', 'Provider Name'])[['Diff b/w Claims and Reimb']].mean().add_prefix("Mean_of_").round(1)
q1d.sort_values(by = ['Mean_of_Diff b/w Claims and Reimb'],ascending = False).head().reset_index()
```

```python
#Part1D) Highest Number of Procedures and Largest Differences between Claims and Reimbursements for out patient
Medicare_Provider_Outpatient['Diff b/w Claims and Reimb'] = Medicare_Provider_Outpatient['Average  Estimated Submitted Charges'] - Medicare_Provider_Outpatient['Average Total Payments']
q1d2 = Medicare_Provider_Outpatient.groupby(by = ['APC', 'Provider Name'])[['Diff b/w Claims and Reimb']].mean().add_prefix("Mean_of_").round(1)
q1d2.sort_values(by = ['Mean_of_Diff b/w Claims and Reimb'],ascending = False).head().reset_index()
```

# Part 2 : Three providers least like others and three regions least like others

```python
#Provider least like others for in patient category
q2a = Medicare_Provider_Inpatient.groupby(by = ['Provider Id'])[['Average Total Payments']].mean().reset_index()
q2a.head()
```

```python
model = LinearRegression()
model.fit(q2a['Provider Id'].values.reshape(-1,1),q2a['Average Total Payments'].values.reshape(-1,1))
```

```python
q2a['Predicted'] = model.predict(q2a['Provider Id'].values.reshape(-1,1))
q2a['Residual'] = abs(q2a['Predicted'] - q2a['Average Total Payments'])
```

```python
#Three Providers least like others for in patient category
q2a.sort_values('Residual',ascending=False).head(3).reset_index()
```
